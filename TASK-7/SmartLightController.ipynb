```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def control_light(self, action):
        if action.upper() == "ON":
            self.status = "ON"
        else:
            self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# User Input
name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(name)
light.control_light(action)
light.display_status()
```
