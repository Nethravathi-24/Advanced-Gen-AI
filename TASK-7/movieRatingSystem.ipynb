```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print(f"Rating: {self.rating} / 5")


# User Input
name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)
movie.display_movie()
```
