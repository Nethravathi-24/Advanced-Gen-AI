```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price(self):
        print("Product:", self.name)
        print(f"Price: ₹{self.price}")


# User Input
name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)
product.display_price()
```
