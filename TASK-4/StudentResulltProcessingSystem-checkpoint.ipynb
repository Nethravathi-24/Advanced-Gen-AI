```python
# Function to calculate average marks and result
def process_result(marks_list):

    total = 0  # variable to store total marks

    # Loop through marks list
    for mark in marks_list:
        total += mark

    # Calculate average
    average = total / len(marks_list)

    # Decide pass or fail
    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"


# Example marks
marks = [60, 70, 55, 40, 65]

# Display result
print(process_result(marks))
```
