```python
# Function to check attendance eligibility
def check_attendance(attendance_list):

    present_days = 0  # counter for present days

    # Loop through attendance list
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100

    # Check eligibility condition
    if percentage >= 75:
        return f"Attendance: {percentage}% - Eligible"
    else:
        return f"Attendance: {percentage}% - Not Eligible"


# Example attendance record
attendance = ["P", "A", "P", "P", "A", "P", "P"]

print(check_attendance(attendance))
```
