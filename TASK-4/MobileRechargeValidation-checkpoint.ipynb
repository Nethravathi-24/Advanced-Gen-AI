```python
# Function to validate mobile recharge amount
def validate_recharge():

    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]

    # Loop runs until user enters a valid recharge
    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Check minimum recharge condition
        if amount < 50:
            print("Minimum recharge amount is ₹50")

        # Check if plan exists
        elif amount not in valid_plans:
            print("Invalid plan selected")

        # Successful recharge
        else:
            print("Recharge Successful!")
            break  # exit loop
            

# Function call
validate_recharge()
```
