```python
# Function to check inventory stock levels
def check_inventory(stock_dict):

    # Loop through dictionary items
    for product, quantity in stock_dict.items():

        # If stock less than 15 â†’ reorder alert
        if quantity < 15:
            print(f"{product}: Reorder Alert")

        # Otherwise stock is sufficient
        else:
            print(f"{product}:Stock OK")


# Example inventory data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

# Function call
check_inventory(inventory)
```
