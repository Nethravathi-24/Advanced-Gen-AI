```python
# Function to calculate cab fare
def calculate_fare():

    base_fare = 50       # fixed starting charge
    rate_per_km = 12     # cost per kilometer

    # Loop allows retry option
    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        # Basic fare calculation
        fare = base_fare + (distance * rate_per_km)

        # Add 25% extra during peak hours
        if peak == "yes":
            fare += fare * 0.25

        print(f"Total Fare: ₹{fare}")

        # Ask user if they want to retry
        retry = input("Calculate again? (yes/no): ").lower()

        if retry != "yes":
            break


# Function call
calculate_fare()
```
