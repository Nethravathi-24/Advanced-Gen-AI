```python
# Function to check password strength
def check_password(password):

    special_chars = "@#$"  # allowed special characters
    has_digit = False
    has_special = False

    # Check minimum length condition
    if len(password) < 8:
        return "Password must be at least 8 characters long"

    # Loop through each character
    for char in password:

        # Check if digit exists
        if char.isdigit():
            has_digit = True

        # Check if special character exists
        if char in special_chars:
            has_special = True

    # Final strength validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (must contain digit & special character)"


# User input
pwd = input("Enter password: ")
print(check_password(pwd))
```
