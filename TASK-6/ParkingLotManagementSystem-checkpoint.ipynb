```python
# Function to manage parking system
def parking_system(capacity, logs):

    parked = 0
    peak_usage = 0

    # Process each entry log
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print(" Over Capacity Alert")
    else:
        print("Parking Available")

    print("Peak Usage:", peak_usage)


# ---- USER INPUT ----
capacity = int(input("Enter parking capacity: "))
log_input = input("Enter logs (IN/OUT separated by space): ").upper().split()

parking_system(capacity, log_input)
```
