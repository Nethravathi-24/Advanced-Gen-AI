```python
# Function to manage registrations
def event_registration(capacity, registrations):

    confirmed = min(capacity, registrations)

    if registrations > capacity:
        waitlist = registrations - capacity
    else:
        waitlist = 0

    print("Confirmed:", confirmed)
    print("Waitlisted:", waitlist)

    if registrations >= capacity:
        print("Registration Closed")
    else:
        print("Registration Open")


# ---- USER INPUT ----
capacity = int(input("Enter event capacity: "))
registrations = int(input("Enter total registrations: "))

event_registration(capacity, registrations)
```
