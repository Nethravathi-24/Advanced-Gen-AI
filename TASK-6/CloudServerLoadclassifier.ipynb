```python
# Function to classify server load
def server_load(cpu_readings):

    total = 0

    # Calculate sum using loop
    for value in cpu_readings:
        total += value

    avg_cpu = total / len(cpu_readings)

    print("Average CPU Load:", avg_cpu, "%")

    # Classification rules
    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)


# Example
server_load([45, 60, 70, 85, 90])
```
