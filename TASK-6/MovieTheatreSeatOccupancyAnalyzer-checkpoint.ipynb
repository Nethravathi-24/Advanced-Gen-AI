```python
# Function to analyze occupancy
def theatre_occupancy(total_seats, booked):

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if occupancy == 100:
        print("Status: Housefull")
    elif occupancy >= 75:
        print("Status: Almost Full")
    else:
        print("Status: Seats Available")


# ---- USER INPUT ----
total_seats = int(input("Enter total seats: "))
booked = int(input("Enter number of booked seats: "))

theatre_occupancy(total_seats, booked)
```
