```python
# Function to monitor resource usage
def classroom_usage(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", overused)

    if overused:
        print("Energy Alert: YES")
    else:
        print("Energy Alert: NO")


# ---- USER INPUT ----
n = int(input("Enter number of resources: "))
resources = {}

for i in range(n):
    name = input("Resource name: ")
    hours = int(input("Usage hours: "))
    resources[name] = hours

classroom_usage(resources)
```
