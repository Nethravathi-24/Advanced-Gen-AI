```python
# Function to estimate delivery time
def delivery_time(distance, traffic, weather):

    time = distance * 5  # base time

    if traffic == "HIGH":
        time += 15
    elif traffic == "MEDIUM":
        time += 8

    if weather == "RAINY":
        time += 10
    elif weather == "STORMY":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# ---- USER INPUT ----
distance = float(input("Enter distance (km): "))
traffic = input("Traffic Level (High/Medium/Low): ").upper()
weather = input("Weather (Clear/Rainy/Stormy): ").upper()

delivery_time(distance, traffic, weather)
```
