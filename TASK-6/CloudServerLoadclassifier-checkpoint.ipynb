```python
# Function to classify server load
def server_load(cpu_readings):

    total = 0

    for value in cpu_readings:
        total += value

    avg_cpu = total / len(cpu_readings)

    print("Average CPU Load:", avg_cpu)

    if avg_cpu < 50:
        print("Status: Normal")
    elif avg_cpu <= 80:
        print("Status: Warning")
    else:
        print("Status: Critical")


# USER INPUT
readings = list(map(int, input("Enter CPU readings separated by space: ").split()))
server_load(readings)
```
